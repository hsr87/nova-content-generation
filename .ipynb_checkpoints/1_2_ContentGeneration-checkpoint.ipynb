```python
import base64
import json
import boto3
import time
client = boto3.client(
    "bedrock-runtime",
    region_name="us-east-1",
)
MODEL_ID = "us.anthropic.claude-3-5-sonnet-20241022-v2:0"

# json_list = [{"schemaVersion":"bedrock-conversation-2024","system":[{"text":"너는 음식 전문가야."}],"messages":[{"role":"user","content":[{"text":"떡볶이의 종류에 대해 알려줘."}]},{"role":"assistant","content":[{"text":"떡볶이의 종류로는 고추장 떡볶이, 간장 떡볶이, 로제 떡볶이, 짜장 떡볶이, 카레 떡볶이, 마라 떡볶이, 국물 떡볶이 등이 있어."}]}]}]
json_list = json_list[0:11]
for i in range(30):
    time.sleep(10)
    messages =[
        {
            "role": "user",
            "content": [
                {
                    "text": """
                                    {
                        "schemaVersion": "bedrock-conversation-2024",
                        "system": [
                            {
                                "text": "너는 음식 전문가야."
                            }
                        ],
                        "messages": [
                            {
                                "role": "user",
                                "content": [
                                    {
                                        "text": "떡볶이의 종류에 대해 알려줘."
                                    }
                                ]
                            },
                            {
                                "role": "assistant",
                                "content": [
                                    {
                                        "text": "떡볶이의 종류로는 고추장 떡볶이, 간장 떡볶이, 로제 떡볶이, \
                                        짜장 떡볶이, 카레 떡볶이, 마라 떡볶이, 국물 떡볶이 등이 있어."
                                    }
                                ]
                            }
                        ]
                    }

                    위와 같은 형식의 한국 음식 관련 데이터 Q&A 데이터를 5건 포함한 하나의 jsonl 파일을 만들어줘. 예시와 똑같은 Q&A 데이터를 생성하진 말아줘. 떡볶이 관련 응답은 1건 이상 있어야 해. 응답 시 jsonl 데이터 외에 다른 응답은 하지 말아줘.
                    """
                }
            ]
        }
    ]
    inf_params = {"maxTokens": 8192, "topP": 1, "temperature": 0.9}
    model_response = client.converse(modelId=MODEL_ID, messages=messages,
                                     inferenceConfig=inf_params)
    json_list.append(model_response['output']['message']['content'][0]['text'])
    
len(json_list)
```

```python
import json

# json_list의 각 항목이 올바른 형식인지 검증하는 함수
def validate_format(item):
    required_keys = {"schemaVersion", "system", "messages"}
    if not isinstance(item, dict):
        try:
            item = json.loads(item)
        except:
            return False
    
    if not all(key in item for key in required_keys):
        return False
    
    if item["schemaVersion"] != "bedrock-conversation-2024":
        return False
    
    if not isinstance(item["messages"], list):
        return False
    
    return True

# 유효한 데이터만 필터링
valid_data = []
for item in json_list:
    if isinstance(item, str):
        try:
            parsed_items = [json.loads(line) for line in item.strip().split('\n') if line.strip()]
            valid_data.extend([item for item in parsed_items if validate_format(item)])
        except:
            continue

# JSONL 파일로 저장
output_file = "korean_food_qa.jsonl"
with open(output_file, 'w', encoding='utf-8') as f:
    for item in valid_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"저장된 유효한 데이터 수: {len(valid_data)}")
```

```python
import base64
import io
import json
from typing import Dict, Any

import boto3
from PIL import Image

def generate_image(model_id: str, body: str) -> bytes:
    bedrock = boto3.client(
        service_name='bedrock-runtime',
        config=Config(read_timeout=300)
    )

    response = bedrock.invoke_model(
        body=body,
        modelId=model_id,
    )
    
    response_body = json.loads(response.get("body").read())
    base64_image = response_body.get("images")[0]
    base64_bytes = base64_image.encode('ascii')
    image_bytes = base64.b64decode(base64_bytes)

    return image_bytes

model_id = 'amazon.nova-canvas-v1:0'
prompt = "Korean tteokbokki - spicy rice cakes in bright red sauce garnished with green onions."

body = json.dumps({
    "taskType": "TEXT_IMAGE",
    "textToImageParams": {
        "text": prompt
    },
    "imageGenerationConfig": {
        "numberOfImages": 1,
        "height": 512,
        "width": 512,
        "cfgScale": 8.0,
        "seed": 0
    }
})

image_bytes = generate_image(model_id=model_id, body=body)
image = Image.open(io.BytesIO(image_bytes))
image.save('food_gen.png',"PNG")
```

```python
import json
import boto3

def generate_video(text: str, duration: int = 6) -> dict:
    bedrock_runtime = boto3.client("bedrock-runtime")
    
    model_input = {
        "taskType": "TEXT_VIDEO",
        "textToVideoParams": {
            "text": text
        },
        "videoGenerationConfig": {
            "durationSeconds": duration,
            "fps": 24,
            "dimension": "1280x720",
            "seed": 0
        }
    }

    response = bedrock_runtime.start_async_invoke(
        modelId="amazon.nova-reel-v1:0",
        modelInput=model_input,
        outputDataConfig={
            "s3OutputDataConfig": {
                "s3Uri": "s3://my-nova-videos-korean-foods"
            }
        }
    )
    
    return response

# Generate video
text_prompt = "steaming hot tteokbokki with rich gochujang sauce and chopped green onions on top"
result = generate_video(text=text_prompt)
print(json.dumps(result, indent=2, default=str))
```

```python
import base64
import io
import json
import logging
from typing import Dict, Any

import boto3
from PIL import Image
from botocore.config import Config
from botocore.exceptions import ClientError


class ImageError(Exception):
    """Custom exception for errors returned by Amazon Nova Canvas."""
    
    def __init__(self, message: str):
        self.message = message


logger = logging.getLogger(__name__)
logging.basicConfig(level=logging.INFO)


def generate_image(model_id: str, body: str) -> bytes:
    """Generate an image using Amazon Nova Canvas model on demand.
    
    Args:
        model_id: The model ID to use
        body: The request body to use
        
    Returns:
        The image generated by the model as bytes
        
    Raises:
        ImageError: If image generation fails
    """
    logger.info("Generating image with Amazon Nova Canvas model %s", model_id)
    
    bedrock = boto3.client(
        service_name='bedrock-runtime',
        config=Config(read_timeout=300)
    )

    accept = "application/json"
    content_type = "application/json"
    
    response = bedrock.invoke_model(
        body=body,
        modelId=model_id,
        accept=accept,
        contentType=content_type
    )
    
    response_body = json.loads(response.get("body").read())
    base64_image = response_body.get("images")[0]
    base64_bytes = base64_image.encode('ascii')
    image_bytes = base64.b64decode(base64_bytes)

    finish_reason = response_body.get("error")
    if finish_reason is not None:
        raise ImageError(f"Image generation error. Error is {finish_reason}")

    logger.info("Successfully generated image with Amazon Nova Canvas model %s", model_id)
    return image_bytes


def main():
    """Entrypoint for Amazon Nova Canvas example."""
    logging.basicConfig(
        level=logging.INFO,
        format="%(levelname)s: %(message)s"
    )

    model_id = 'amazon.nova-canvas-v1:0'
    prompt = "Korean tteokbokki - spicy rice cakes in bright red sauce garnished with green onions."
    
    body = json.dumps({
        "taskType": "TEXT_IMAGE",
        "textToImageParams": {
            "text": prompt
        },
        "imageGenerationConfig": {
            "numberOfImages": 1,
            "height": 1024,
            "width": 1024,
            "cfgScale": 8.0,
            "seed": 0
        }
    })

    try:
        image_bytes = generate_image(model_id=model_id, body=body)
        image = Image.open(io.BytesIO(image_bytes))
        image.show()
        
    except ClientError as err:
        message = err.response["Error"]["Message"]
        logger.error("A client error occurred: %s", message)
        print(f"A client error occurred: {message}")
        
    except ImageError as err:
        logger.error(err.message)
        print(err.message)
        
    else:
        print(f"Finished generating image with Amazon Nova Canvas model {model_id}.")


if __name__ == "__main__":
    main()
```
